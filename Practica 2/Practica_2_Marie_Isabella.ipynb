```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from statistics import mean
```

Prueba del funcionamiento de la función para el Impulso unitario




```python
def impseq(n0, n1, n2):
# Genera x(n) = delta(n-n0); n1 <= n <= n2
# ----------------------------------------------
 n_i = np.arange(n1,n2+1) # Se crea el vector de muestras
 x_i = (n-n0) == 0
 return [x_i,n_i]
```

```python
x,n = impseq(5,1,10)
plt.stem(n,x)
plt.grid(True)
plt.title("Impulso")
```

```python
def escalon(n0,n1,n2):
  n = np.arange(n1,n2+1) # Se crea el vector de muestras
  x=(n-n0) >= 0
  return [x,n]
```

```python
x,n = escalon(5,1,10)
plt.stem(n,x)
plt.grid(True)
plt.title("Escalón")
```

image.png

```python
def rampa(n0,n1,n2):
  n = np.arange(n1,n2+1) # Se crea el vector de muestras
  u,n = escalon(n0,n1,n2)
  x = (n-n0)*u
  return [x,n]
```

```python
x,n = rampa(5,1,10)
plt.stem(n,x)
plt.grid(True)
plt.title("Rampa")
```
