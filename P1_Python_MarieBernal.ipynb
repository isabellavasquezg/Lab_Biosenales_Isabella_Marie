Escribir una funci√≥n que reciba un diccionario con las notas de los alumnos de un curso y devuelva una
serie con la nota m√≠nima, la m√°xima, media y la desviaci√≥n t√≠pica.

```python
import pandas as pd
dic_notas = {"Marie": 3.8, "Laura": 4.8, "Dairon": 5.0, "Anny": 4.6, "Paula": 4.0, "Paulina": 5.0, "Andrea": 4.5}
def cursobio(dic_notas):
  notas = list(dic_notas.values())
  min = np.min(notas)
  max = np.max(notas)
  media = np.mean(notas)
  sigma = np.std(notas)
  result_serie = pd.Series([min, max, media, sigma], index=['nota m√≠nima', 'nota m√°xima', 'media', 'desviaci√≥n t√≠pica'])
  return result_serie
print (result_serie)
```
