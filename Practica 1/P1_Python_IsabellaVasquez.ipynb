1. **np.ones**

Devuelve una nueva matriz de la forma y tipo dados, rellena con unos.

2. **np.round**

Devuelve un array equivalente al cedido como argumento tras sustituir cada elemento por la versión redondeada a un cierto número de decimales.

3. **np.ceil**

Devuelve el menor entero mayor o igual que los elemento de la estructura de entrada

4. **np.floor**

Devuelve el mayor entero menor o igual que los elementos de la estructura de entrada




---


Referencia: [Numpy](https://numpy.org/)

```python
print("Ejemplo del uso de np.ones:")
Munos = np.ones((3,3), dtype=int)
print("Matriz de unos 3x3 creada a partir del comando: \n ",Munos)
print('\n')

print("Ejemplo del uso de np.round:")
a_round = np.round(a, 2)
print("Vector a redondeando sus elementos en 2 cifras significativas:  ", a_round)
print('\n')


print("Ejemplo del uso de np.ceil:")
b_ceil = np.round(b)
print("Vector b modificado para que se muestre solo el menor entero mayor o igual de sus elementos:  ", b_ceil)
print('\n')

print("Ejemplo del uso de np.floor:")
ab_floor = np.floor(a*b)
print("Vector a*b modificado para que se muestre solo el menor entero mayor o igual de sus elementos:  ", ab_floor)
print('\n')
```

```python
#n = np.arange(0,81,1) #se genera el vector que va de 0 a 80 con un paso de 1
n = np.linspace(0,80,80)
y_n = np.sin((np.pi)*(0.18)*n)
y_2n = np.cos((2)*(np.pi)*(0.03)*n)
s_n = y_n + y_2n
t_n = y_n*y_2n
```

```python
plt.subplots_adjust(hspace=0.5, wspace=0.6)

plt.subplot(2,1,1)
plt.plot(n, y_n, "#ff8033")
plt.title('Señal y[n]')
plt.xlabel('n')
plt.ylabel('y[n]')
plt.grid(True)

plt.subplot(2,1,2)
plt.plot(n, y_2n,"#33ff3f")
plt.title('Señal y2[n]')
plt.xlabel('n')
plt.ylabel('y2[n]')
plt.grid(True)
plt.show()
```

```python
plt.subplots_adjust(hspace=0.5, wspace=0.6)

plt.subplot(2,1,1)
plt.plot(n, s_n, "#33e0ff")
plt.title('Señal s[n] = y[n] + y2[n]')
plt.xlabel('n')
plt.ylabel('s[n]')
plt.grid(True)

plt.subplot(2,1,2)
plt.plot(n, t_n,"#9933ff")
plt.title('Señal t[n] = y[n] * y2[n]')
plt.xlabel('n')
plt.ylabel('t[n]')
plt.grid(True)
plt.show()
```

```python
plt.plot(n, y_n, "#ff8033", label='y[n]')
plt.plot(n, y_2n, "#33ff3f", label='y2[n]')
plt.xlabel('n')
plt.ylabel('y[n] --- y2[n]')
plt.grid(True)
plt.legend()
plt.title("Señales y[n] y y2[n]")
```

```python
plt.plot(n, t_n, "#9933ff", label='t[n]')
plt.plot(n, s_n, "#33e0ff", label='s[n]')
plt.xlabel('n')
plt.ylabel('s[n] -- t[n]')
plt.grid(True)
plt.legend()
plt.title("Señales t[n] y s[n]")
```

```python
notas_alumnos = { "Julian": 3.0, "Sandra": 4.5, "Diana": 5.0, "Lorena": 3.0, "Simon": 4.54, "Sebastian": 5.0, "Maria": 3.33,
                 "Eva": 4.56, "León": 5.0,"Luis": 3.76, "Marco": 4.8, "Sofía": 3.0, "Manuel": 1.0,
                  "Arturo": 4.5, "Andrés": 5.0, "Ana": 3.9, "Luisa": 4.8, "Miguel": 0.9}

print("Diccionario Notas Alumnos:\n",notas_alumnos)

def serie(dict):
  serie_notas_alumnos = pd.Series(dict)
  max = serie_notas_alumnos.max()
  min = serie_notas_alumnos.min()
  media = serie_notas_alumnos.mean()
  dt = serie_notas_alumnos.std()
  return print("\nInformación estadística: \nNota Máxima {} \nNota Minima {} \nMedia {} \nDesviación estándar {} \n" .format(max, min, media, dt))

serie(notas_alumnos)
```
